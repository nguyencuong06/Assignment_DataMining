## 2) Data preparation and preprocessing

### Load data

Data loaded by using Python, using Pandas library, especially: pd.read_csv('data.csv')

### Normalize, Clean, Transform data

```python
# Step 1: Remove column Id, because we don't need it, Ids do not have any value but indexes.
del data['Id']
data.head(10)
```

```python
# Step 2: Normalize column Date and convert its type to datetime for later convenience
data['Date']=data['Date'].str.split('T').str.get(0)
data['Date']=pd.to_datetime(data.Date)
data.head(2)
```

```python
# Step 3: Fill missing values
data=data.fillna('NA')
data.count()
```

```python
# Step 4: Remove(Clean) HTML tags for 'FullDescription'
```

```python
data[data.FullDescription!='NA'].head(1)

# before
```

```python
from bs4 import BeautifulSoup
data['FullDescription'] = [BeautifulSoup(text).get_text() for text in data['FullDescription']]
data[data.FullDescription!='NA'].head(1)

# after
```

> The results are not so good but quite acceptably.

```python
# Step 5: Remove(Clean) emoji from 'Title'
```

```python
# emoji samples from data
(data[data['Title'].str.contains('👑')])

# before
```

```python
# pip install emoji
import emoji
data['Title'] = [emoji.get_emoji_regexp().sub(r'', text) for text in data['Title']]
```

```python
data.loc[[693,203134,231630]]

# after
```

```python
# Step 6: Reorder columns
data=data[['Title', 'Company', 'Date', 'Location', 'Area', 'Classification',
       'SubClassification', 'Requirement', 'FullDescription','JobType','LowestSalary',
       'HighestSalary']]
data[(data.Area!='NA') & (data.FullDescription!='NA')].head(2)
```

```python
# Step 7: Save as new file (edited file) for later use(s)
data.to_csv('data_edited.csv',index=False)
```
