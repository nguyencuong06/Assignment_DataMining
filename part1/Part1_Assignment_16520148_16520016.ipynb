```python
#Apply Min-Max Scaling to LowestSalary and HighestSalary column
df['LowestSalary'] = (df['LowestSalary'] -( df['LowestSalary'].min()))/ ( (df['LowestSalary'].max()) - (df['LowestSalary'].min()))
df['HighestSalary'] = (df['HighestSalary'] - (df['HighestSalary'].min()))/ ( (df['HighestSalary'].max()) - (df['HighestSalary'].min()))
```

##NOTICED : Why does Id contain string ?

```python
#Check for Nan value 
df.isnull()
```

```python
#Delete row which contain null
df = df.dropna(how = 'any').reset_index()
```

```python
#Check null value again
df.isnull()
len(df)
```

```python
#Check for Id column. We found that column Id contain string('&....').We start to clean it
df['Id'].value_counts()
```

```python
#Find which Id's Index contain &..... Value > -1 mean that it contains the string (&...)
x = df['Id'].str.find('&')
```

```python
#Check which index that value contain string (&....) We see that range start with index = 90060 will contains string (&...)
x[x>-1]
```

```python
#Split the value with & 
df['Id'][90060:] = df['Id'][90060:].str.split('&')
```

```python
#Get value before & mean you get Id number only
df['Id'][90060:] = df['Id'][90060:].iloc[0][0]
```

```python
#Check for outlier value ( '&...')
df['Id'].str.contains('&')
```
